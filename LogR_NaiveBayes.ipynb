```python
data_train['label'][data_train['label']!='normal.'] = 'attack'  # end with .
data_train['label'][data_train['label']=='normal.'] = 0
data_train['label'][data_train['label']=='attack'] = 1
data_train[['label']] = data_train[['label']].apply(pd.to_numeric)
data_train['label'].value_counts()
```

```python
data_test['label'][data_test['label']!='normal.'] = 'attack'  # end with .
data_test['label'][data_test['label']=='normal.'] = 0
data_test['label'][data_test['label']=='attack'] = 1
data_test[['label']] = data_test[['label']].apply(pd.to_numeric)
data_test['label'].value_counts()
```

#########################################Above is preprocessing test #########################################################

```python
x_train = pd.get_dummies(data_train.drop('service', axis=1).iloc[:, 0:40])
y_train = data_train['label']
x_test = pd.get_dummies(data_test.drop('service', axis=1).iloc[:, 0:40])
y_test = data_test['label']
```

##### Get trainning and test data, 'service' column is dropped.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model.logistic import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

LogisticRegression(penalty='l1',
                            dual=False,
                            tol=1e-4,
                            C=1.0,
                            fit_intercept=True,
                            intercept_scaling=1,
                            class_weight=None,
                            random_state=None,
                            solver='liblinear',
                            max_iter=100,
                            multi_class='ovr',
                            verbose=0,
                            warm_start=False,
                            n_jobs=1)
                            
### Note that this is default setting for the logistic regression. We would test different iterations to see the differences

```python
for i in range(30):
    iteration = i*10
    model = LogisticRegression(max_iter=iteration, solver='lbfgs')
    model.fit(x_train, y_train)
    predictions=model.predict(x_test)
    print("Iteration: ", iteration, " Accuracy score:", accuracy_score(y_test,predictions))
```

### Note: the iteration at 90 reaches stable accuracy, but after 210 iterations, it is stable

```python
model = LogisticRegression(max_iter=90)
model.fit(x_train, y_train)
predictions=model.predict(x_test)
print(classification_report(y_test, predictions))
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
import matplotlib.pyplot as plt
```

```python

fpr, tpr, threshold = roc_curve(y_test, predictions)
roc_auc = auc(fpr, tpr)

plt.title('Receiver Operating Characteristic Curve')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
```

```python
NB_G_model = GaussianNB()
NB_M_model = MultinomialNB()
NB_B_model = BernoulliNB()
```

```python
NB_G_model.fit(x_train, y_train)
NB_M_model.fit(x_train, y_train)
NB_B_model.fit(x_train, y_train)
```
